```python
# Setting customer_id to be the Index
df = pd.read_csv('DM2425_ABCDEats_DATASET.csv', sep=',', index_col='customer_id') 
df.head()
```

```python
# Checking the data types of our features and potential missing values
df.info() 
```

```python
# Finding all the customer_id that are duplicated
duplicate_customer_ids = df.index[df.index.duplicated()].unique()  
print(duplicate_customer_ids)
```

```python
# Dropping duplicated customer_id values, keeping only the first occurrence
df = df[~df.index.duplicated(keep='first')]
```

```python
# Checking if there are still Duplicated Values
df.duplicated().sum()
```

```python
# Understanding which rows have duplicated data, but different customer_id
duplicates_with_different_ids = df[df.duplicated(subset=df.columns.difference(['customer_id']), keep=False)]
print(duplicates_with_different_ids)
```

```python
# Removing rows with duplicated data (but with different customer_id)
df = df.loc[~df.duplicated(subset=df.columns.difference(['customer_id']))]
```

<a id="hr-0"></a>
### **HR_0**

```python
# Considering all values in HR_0 column as missing values
df['HR_0'] = df['HR_0'].replace(0, np.nan)

# Sum of columns from DOW_0 to DOW_6
dow_sum = df.loc[:, 'DOW_0':'DOW_6'].sum(axis=1)
# Sum of columns from HR_0 to HR_23 
hr_sum = df.loc[:, 'HR_0':'HR_23'].sum(axis=1)

# Calculating the difference: dow_sum - hr_sum
difference = dow_sum - hr_sum

# Filling missing values in HR_0 with the calculated 'difference'
df['HR_0'].fillna(difference, inplace=True)
```

```python
df['HR_0'].isna().sum()
```

```python
# Selecting columns representing orders per hour (HR_0 to HR_23)
orders_per_hour = df.loc[:, 'HR_0':'HR_23']

# Calculating the total number of orders for different times of the day
dawn = sum(df.loc[:, 'HR_0':'HR_6'].sum())  # Sum of orders from 12 AM to 6 AM
morning = sum(df.loc[:, 'HR_7':'HR_12'].sum())  # Sum of orders from 7 AM to 12 PM
afternoon = sum(df.loc[:, 'HR_13':'HR_18'].sum())  # Sum of orders from 1 PM to 6 PM
night = sum(df.loc[:, 'HR_19':'HR_23'].sum())  # Sum of orders from 7 PM to 11 PM

# Organizing total orders for each time of the day in a list
orders_per_moment_of_day = [dawn, morning, afternoon, night]

# Selecting columns representing orders per day of the week (DOW_0 to DOW_6)
orders_per_dow = df.loc[:, 'DOW_0':'DOW_6']

# Calculating the sum of orders for each day of the week
orders_per_dow_sum = orders_per_dow.sum()

# Selecting columns representing the amount spent per cuisine type (CUI_American to CUI_Thai)
amount_spent_per_cuisine = df.loc[:, 'CUI_American':'CUI_Thai']

# Calculating the sum of the amount spent for each cuisine type
amount_spent_per_cuisine_sum = amount_spent_per_cuisine.sum()
```

```python
# Splitting data into metric and non metric
df_non_metric = df.select_dtypes(include=['object'])
df_metric = df.select_dtypes(include=['int64', 'float64'])
```

## Grouped Metric Features

### Days of the Week 

```python
# Calculating the average number of orders per day of the week
average_orders_dow = orders_per_dow_sum.mean()

# Creating a list with the days of the week 
list_day_of_week = ['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday']

plt.figure(figsize=(10, 6))

# Plotting the total orders per day of the week as a bar chart
orders_per_dow_sum.plot(kind='bar', color = 'peru')
plt.xticks(ticks=range(len(list_day_of_week)), labels=list_day_of_week, rotation=0)
plt.axhline(y=average_orders_dow, color='sienna', linestyle='--', label='Average Orders')
plt.title('Total Orders per Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Total Orders')
plt.xticks(rotation=0)
plt.legend()
plt.show()

```

### Days of the Week - Remarks
- The total number of orders per day of the week does **not have a high disparity**, as we can observe in the graph;
- We decided to compare it to the mean, approximately **19,902** orders per day of the week. This comparison allows us to identify days that **fall below it and those that meet or exceed it**;
- For example, from **Thursday to Saturday**, the number of orders **exceeds** the average, while the rest of the days **fall below it**. Unexpectedly, **Sunday has the lowest** order count of the week.

### Moments of the day

```python
# Creating a list with the moments of the day
moments_of_day = ['Dawn', 'Morning', 'Afternoon', 'Night']

# Saving the average number of orders per moment of the day, respectively
orders_per_moment_of_day = [15.547, 46.632, 57.712, 17.611] 

# Calculating the average number of orders per moment of the day
average_orders_moment = sum(orders_per_moment_of_day) / len(orders_per_moment_of_day)

# Calculating the angles for the radar chart
angles = np.linspace(0, 2 * np.pi, len(moments_of_day), endpoint=False).tolist()

# Completing the loop for the radar chart by appending the first value to the end
orders_per_moment_of_day += orders_per_moment_of_day[:1] 
angles += angles[:1] 

# Polar subplot
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))

# Plotting the total orders per moment of the day
ax.plot(angles, orders_per_moment_of_day, color='peru', linewidth=2, label='Total Orders')

# Filling the area under the plot
ax.fill(angles, orders_per_moment_of_day, color='peru', alpha=0.3)

# Plotting the average orders as a dashed line
ax.plot(angles, [average_orders_moment] * len(angles), color='sienna', linestyle='--', label='Average Orders')

# x-ticks to represent the moments of the day
ax.set_xticks(angles[:-1])  
ax.set_xticklabels(moments_of_day)  

# Setting y-ticks and labels
ax.set_yticks([15, 30, 45, 60])  
ax.set_yticklabels(['15', '30', '45', '60'], color="grey", size=10) 
 
plt.title('Total Orders per Moment of the Day', size=16, y=1.1)
plt.legend(loc='upper right', bbox_to_anchor=(1.1, 1.1))
plt.show()
```

### Moments of the Day - Remarks
- We defined each time segment as follows: **Dawn (midnight to 6a.m.), Morning (7a.m. to 12p.m.), Afternoon (1p.m. to 6p.m.), and Night (7p.m. to 11p.m.)**;
- The values used on the graph are **in thousands**;
- Once again, we used the **average** to compare with the different moments of the day;
- With a mean of **34,440** per moment of the day, both **morning (46,632 orders)** and **afternoon (57,712 orders)** show **higher** order volumes, with the **afternoon** having the **highest** order count;
- As expected, order counts are **low during dawn (15,547 orders)**, however, they are also **unexpectedly low at night (17,611 orders)**.

## customer_lifecycle_stage

```python
# Defining the thresholds for customer lifecycle stages
recent_threshold = 30
long_term_threshold = 30

# Calculating the number of days since last order
df['days_since_last_order'] = df['last_order']

# Defining the customer lifecycle stage function
def assign_lifecycle_stage(row):
    # New customer: if the first order was in the last recent_threshold days
    if row['first_order'] > recent_threshold:
        return 'New'
    # Active customer: if the customer has been around for a longer time but still active
    elif row['days_since_last_order'] <= long_term_threshold:
        return 'Active'
    # Lapsed customer: if the customer hasn't ordered in a while
    else:
        return 'Lapsed'

# Applying the function to assign lifecycle stages
df['customer_lifecycle_stage'] = df.apply(assign_lifecycle_stage, axis=1)

# To see the results along with the customer_id, we can reset the index temporarily
# or use .index to display the customer IDs along with lifecycle stage

# Option 1: Reset the index temporarily to include 'customer_id'
df_reset = df.reset_index()

# Option 2: Access the index directly if you don't want to reset it
df['customer_lifecycle_stage'] = df.apply(assign_lifecycle_stage, axis=1)
df_reset[['customer_id', 'customer_lifecycle_stage']].head()

# Viewing the results (if you need to use customer_id as part of the DataFrame)
print(df_reset[['customer_id', 'customer_lifecycle_stage']].head())
```

```python
# Plotting the distribution of customer lifecycle stages
plt.figure(figsize=(8, 6))
sns.countplot(x='customer_lifecycle_stage', data=df, palette='winter')
plt.title('Customer Lifecycle Stage Distribution')
plt.xlabel('Lifecycle Stage')
plt.ylabel('Number of Customers')
plt.show()
```

```python
# Analyzing the distribution of customer lifecycle stages
df['customer_lifecycle_stage'].value_counts()
```

### Customer Lifecycle Stages - Remarks
- This bar plot shows how many days, **on average**, customers in each lifecycle stage have since their last order. You can use this to analyze the retention of new versus lapsed customers;
- **Lapsed** is the **largest** group, representing customers who have **not placed an order in a long time**. **Re-engagement** efforts could help bring some of these **15,529** customers back;
- The **12,471** New customers show strong growth in customer acquisition. **Retention** strategies could ensure they transition into long-term loyal customers;
- The **3,828** Active customers are the most valuable group, as they regularly place orders. **Loyalty** programs or rewards can help maintain their engagement and spending.
